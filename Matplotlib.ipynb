```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,20,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,50,60,80,85,90,70,75,80,85]
```

```python
numpyyasListesi = np.array(yasListesi)
```

```python
numpykiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyyasListesi,numpykiloListesi,"purple")
plt.xlabel("Yaş Listesi")
plt.ylabel("Kilo Listesi")
```

```python
numpyDizisi1 = np.linspace(0,11,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1**2
```

```python
numpyDizisi3 = numpyDizisi1**3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyDizisi1,numpyDizisi3,"g*-")
```

```python
myfigure = plt.figure()
figureAxes = myfigure.add_axes([0.2,0.2,1,1])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni")
figureAxes.set_ylabel("Y Ekseni")
figureAxes.set_title("Grafik")
```

```python
myfigure = plt.figure()
eksen1 = myfigure.add_axes([0.2,0.2,2,2])
eksen2 = myfigure.add_axes([0.2,0.2,2,2])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Grafik")

eksen2.plot(numpyDizisi1,numpyDizisi3,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Grafik")
```
